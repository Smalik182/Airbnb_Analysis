# Airbnb in Austin

We are going to try to figure out where to purchase property to operate an Airbnb in Austin, Texas. Austin is a booming city that has seen tremendous growth in the past few years. Many are looking to invest in the city and today we are going to attempt to find out exactly what areas we should be looking at. 

We are working with a data set from InsideAirbnb.com 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


# Read in the data
airbnb = pd.read_csv('listings.csv')

# Quick exploration of data
print(airbnb.shape)
pd.options.display.max_columns = 150 # to avoid truncated output 
pd.options.display.max_rows = 50
airbnb.head()
```

## Data Cleaning

We should check for any errors with the data types of the columns. We can see that the 'Price' column is a string data type due to the '$'. Lets fix that

```python
airbnb.info()
```

```python
airbnb['price'] = airbnb['price'].replace({'\$' : '', ',' : ''} , regex = True).astype(float)
airbnb['price'].info()
```

## Projected Revenue on Airbnbs

We can get a relative idea of the earnings on these properties. Since there is no direct column with the earnings we will have to create one. 

We can use the 'availability_30' to get an idea of how many days a airbnb is booked every month. And the 'price' to calculate the earning for the month.

```python
airbnb['proj_rev_30'] = (30 - airbnb['availability_30']) * airbnb['price']
airbnb['proj_rev_30'].head()
```

## What are the highest earning Airbnbs in the city

Now that we have a relative idea of the projected revenue, lets see if we can get an idea on some of the highest earners. Let's take a look at the 10 highest earners.

```python
highest_earners = airbnb.nlargest(n=10, columns = 'proj_rev_30')
highest_earners.head()
```

We can can see the data does not look right. The airbnbs are fully booked at extremely high prices, this is highly unlikely. After going to the URL's of some of these postings, we discovered that they were at much lower costs or were not up at all.

- First posting is not up at all
- Second posting is listed for $84
- Third posting is listed for $55

We should change our approach. Let's try to work with the 'number_of_reviews' column to get a more accurate data of Airbnbs that are actually being booked.

## Host Location

We can see that for some reason the data set contains some airbnbs that are located in 'San Francisco, CA'. After further inspection we found out that there are some random data from other cities as well. We need to remove these from our data set.

We are then going to try to figure out the top 500 'reviews_per_month' Airbnbs.

```python
# Remove any rows where 'host_location' != 'Austin, Tx'
clean_location = airbnb[airbnb["host_location"].str.contains("Austin, Tx")==False]

# Get 500 most popular airbnbs
popular_airbnb = clean_location.nlargest(n=500, columns = 'number_of_reviews')
popular_airbnb
```

## Where are the most popular Airbnbs in the city?

Now that we have a clearer picture of the most popular Airbnbs, lets try and get some insight as to where they generally are. 

We can see that the data includes zipcodes for the Airbnbs. We can use this to visualize the frequency of the zip codes. 

```python
airbnb_zip = popular_airbnb['neighbourhood_cleansed'].value_counts()
airbnb_zip
```

```python
# Function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center', fontsize = 8)

x = ['','','','','','','','','', '','','','', '']
y = [139, 111,29,28,20,16,15,15,14,10,10,9,7, 7]        
        
# Graph for Frequency Table above
plt.style.use('fivethirtyeight')

airbnb_zip.plot.bar()
plt.title('Popular Airbnb by Zipcodes',
          y = 1.08) # y pads the title upward'
plt.ylabel('Count', fontsize = 12)
plt.xticks(rotation = 90)
plt.ylim([0,150])
# calling the function to add value labels
addlabels(x, y)
plt.show()
```

We can see from the frequency table that users are booking the most airbnbs in:
- Central East Austin (78702) ~ around 27.8%
- South River City (78704) ~ around 22.2%
- Congress Ave District (78701) ~ around 5.8%

# Lets see some outliers on the Price


First lets try to filter Airbnbs that have actually been booked in the past. We can do this by working with the 'number_of_reviews' column

```python
zip_code_5 = popular_airbnb.loc[(popular_airbnb['neighbourhood_cleansed'] == 78702) |
                           (popular_airbnb['neighbourhood_cleansed'] == 78704) |
                           (popular_airbnb['neighbourhood_cleansed'] == 78701) |
                           (popular_airbnb['neighbourhood_cleansed'] == 78744) |
                           (popular_airbnb['neighbourhood_cleansed'] == 78721)]


zip_code_5['neighbourhood_cleansed'].value_counts()

```

```python
sns.boxplot(y = 'price', x = 'neighbourhood_cleansed',  data = zip_code_5)
plt.title('Cost of Airbnbs by Area',
         fontsize = 16)

plt.ylabel('Price of Airbnb')
plt.xlabel('Country')
plt.xticks(range(5), ['78702', '78704', '78701', '78721', '78744']) # avoids tick labels overlap
plt.show()
```

We can see that 78701 contains an outlier that is extremely high compared to the rest, lets try to figure out what is going on.

```python
#zip_code_5 = zip_code_5[zip_code_5['price'] < 20000]
expensive = zip_code_5[zip_code_5['price'] > 8000]
expensive
```

We can see that the price is mostly an error at '$9999.0' while only being a 2 bedroom and 1 bath Airbnb. We should remove this outlier.

```python
cleaned_price = zip_code_5[zip_code_5['price'] < 8000]
```

Now lets plot the boxplot again.

```python
sns.boxplot(y = 'price', x = 'neighbourhood_cleansed',  data = cleaned_price)
plt.title('Cost of Airbnbs by Area',
         fontsize = 16)

plt.ylabel('Price of Airbnb')
plt.xlabel('Country')
plt.xticks(range(5), ['78702', '78704', '78701', '78721', '78744']) # avoids tick labels overlap
plt.show()
```

We can see above by the boxplot that Central East Austin (78702) has the most range in price for Airbnbs. This makes sense since Central East Austin is a very centralized location with many attractions in the area. 

South River City (78704) offers less of a range in pricing and is cheaper, but we can note that there are some more expensive Airbnbs in the area. 

Congress Ave District (78701) seems to have more of a range in pricing than South River City. As well as more expensive Airbnbs. 


## Around what price range are people booking Airbnbs?

Now that we have an idea of generally where most Airbnbs are, lets try to figure out some price ranges that people are booking the most popular Airbnbs at.

```python
ax = cleaned_price['price'].sort_values().plot.bar()
plt.title('Price Range Across popular Airbnbs',
          y = 1.08) # y pads the title upward'
plt.ylabel('Price', fontsize = 12)
plt.xticks([])
plt.ylim([0,800])
plt.show()
```

We can see that majority of Airbnbs are under \$200 a night but lets go a bit further and try to seperate the data by:
- Less than $100
- Greater than $100 but less than $200
- Greater than $200

```python
under_100 = cleaned_price[cleaned_price['price'] <= 100 ]
between_100_200 = cleaned_price[(cleaned_price['price'] >= 100) & (cleaned_price['price'] <= 200) ]
above_200 = cleaned_price[cleaned_price['price'] > 200 ]
above_200
```

There are 125 Airbnbs under $100
There are 134 Airbnbs from $100 to $200
There are 44 Airbnbs above $200


```python


plt.figure(figsize = (10,5))
# first subplot
plt.subplot(1, 3, 1)
under_100['proj_rev_30'].sort_values().plot.bar()
plt.title('Under $100', fontsize = 12) # y pads the title upward'
plt.xticks([])

# Second subplot
plt.subplot(1, 3, 2)
between_100_200['proj_rev_30'].sort_values().plot.bar()
plt.title('Between \$100 and \$200', fontsize = 12)
plt.xticks([])

# Third subplot
plt.subplot(1, 3, 3)
above_200['proj_rev_30'].sort_values().plot.bar()
plt.title('Above $200', fontsize = 12)
plt.xticks([])
plt.ylim(0,20000)

# Show
plt.tight_layout()
plt.show()
```

This gives us a good idea of what we can expect as income off of an airbnb at the different price ranges. Again this is using the projected 30 day revenue we calculated by:
- (30 - 'availability_30') * 'price'
